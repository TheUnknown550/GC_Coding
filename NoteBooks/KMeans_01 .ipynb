# ทดลองทำ Clustering : K-Means #1

---

สร้างข้อมูล (โดยใช้ sklearn datasets .. make blobs) 

*   กำหนด n_samples (จำนวน Samples) = 2000 จุด
*   กำหนด random_state เพื่อให้การสุ่มครั้งต่อไปเหมือนเดิม


```python
from sklearn.datasets import make_blobs
```

```python
n_samples = 2000
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)
```

เรียกใช้ Library Matplotlib .. เพื่อแสดงผลข้อมูล

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0], X[:,1])
plt.rc('xtick', labelsize=18)    # fontsize of the tick labels
plt.rc('ytick', labelsize=18)    # fontsize of the tick labels
plt.show()
```

จาก Library sklearn (หรือ scikit learn) เราจะเรียกใช้ Cluster .. KMeans

```python
from sklearn.cluster import KMeans
```

ทำการ Clustering โดยระบุค่า K=3

```python
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)
```

แสดงผลการ Clustering ด้วย K Means (K=3)

```python
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.rc('xtick', labelsize=18)    # fontsize of the tick labels
plt.rc('ytick', labelsize=18)    # fontsize of the tick labels
plt.title("K Means")
```



---



หากลองเปลี่ยน ค่า K




---



ทดลอง ทำ Clustering อีกครั้ง โดย ระบุค่า K=4

```python
y_pred_k4 = KMeans(n_clusters=4, random_state=random_state ).fit_predict(X)
```

แสดงผลการ Clustering ด้วย K Means (K=4)

```python
plt.scatter(X[:, 0], X[:, 1], c = y_pred_k4)
plt.rc('xtick', labelsize=18)    # fontsize of the tick labels
plt.rc('ytick', labelsize=18)    # fontsize of the tick labels
plt.title("K Means : Incorrect K" )
```
